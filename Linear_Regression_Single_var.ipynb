```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn import metrics
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
```

```python
data.head()
```

```python
data
data.shape
```

```python
X = data.drop(columns = 'per capita income (US$)')
y = data['per capita income (US$)']
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
plt.scatter(X,y,)
plt.title("Canada_per_capita_income")
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 0)
```

```python
lr = linear_model.LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
print(f"Intercept is : {lr.intercept_}")
print(f"Coefficient value is : {lr.coef_}")
```

```python
y_pred  = lr.predict(X_test)
```

```python
plt.scatter(X,y,)
plt.title("Canada_per_capita_income")

plt.xlabel("year")
plt.ylabel("per capita income (US$)")


plt.plot(X_test,y_pred)

plt.show()
```

```python
#Calculation of evaluation metrics mean squared error and r2_value
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("R2 score of the linear regression model is : ",r2_score(y_test,y_pred))
print("MSE of the linear regression model is : ",mean_squared_error(y_test,y_pred))
```
